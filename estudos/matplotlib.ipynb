# Plotar Reta 2D e ponto

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = -1
b = 4
c = 0.4

x = np.linspace(-2, 4, 20)
y = (-a*x - c) / b

# reta
plt.plot(x, y)
plt.grid(True)
plt.axhline(0, -2, 4, color='k', linewidth=1)
plt.axvline(0, -1, 1, color='k', linewidth=1)
plt.title('Reta 2D e ponto')
plt.xlabel('x')
plt.ylabel('f(x)')

# ponto
p = (0, -0.1)
ret = a * p[0] + b * p[1] + c
print('ret = %.2f' % ret)
plt.plot(p[0], p[1], color='r', marker='o', markersize=7)
```

# Plotar m√∫ltiplos gr√°ficos - Subplot

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([-2, -1, 0, 1])
y1 = 2 * x1

x2 = np.array([-2, -1, 0, 1])
y2 = np.array([1, 2, 0, 1])

plt.figure(figsize=(15, 4))

# plot 1
plt.subplot(1, 2, 1) # linhas, colunas, index
plt.plot(x1, y1)
plt.title('Gr√°fico 1')

# plot 2
plt.subplot(1, 2, 2) # linhas, colunas, index
plt.plot(x2, y2)
plt.title('Gr√°fico 2')

plt.suptitle('Gr√°ficos')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([-2, -1, 0, 1])
y1 = 2 * x1

x2 = np.array([-2, -1, 0, 1])
y2 = np.array([1, 2, 0, 1])

# plot 1
plt.subplot(2, 1, 1) # linhas, colunas, index
plt.plot(x1, y1)
plt.title('Gr√°fico 1')

plt.subplots_adjust(hspace=0.5) # espa√ßo entre os gr√°ficos

# plot 2
plt.subplot(2, 1, 2) # linhas, colunas, index
plt.plot(x2, y2)
plt.title('Gr√°fico 2')

plt.show()
```
